Generate a random number.

```python
"""Dummy notebook."""
import random

import matplotlib.pyplot as plt
import numpy as np

# Ensures random numbers generated are the same every run
np.random.seed(42)
random.seed(42)

new_number = np.random.randint(1, 10)
print(new_number)

```

Generate a graph.

```python
# Sample data
x = [1, 2, 3, 4, 5]
y = [10, 20, 25, 30, 20]

# Create a line plot
plt.plot(x, y, marker='o')
plt.title('Sample Line Plot')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.grid()
plt.show()
```
